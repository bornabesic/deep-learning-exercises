## 3. Vector Norm



**(a)**




**i)**
\begin{equation}
\mathbf{x_1} = \begin{bmatrix}
    24\\
    3\\
    2\\
    31\\
\end{bmatrix}
\end{equation}

\begin{equation}
\mathbf{||x_1||_1} = 24+3+2+31 = 60
\end{equation}

\begin{equation}
\mathbf{||x_1||_2} = \sqrt{24^2+3^2+2^2+31^2} = 39.37
\end{equation}

\begin{equation}
\mathbf{||x_1||_8} = (24^8+3^8+2^8+31^8)^\frac{1}{8} = 31.47
\end{equation}

\begin{equation}
\mathbf{||x_1||_\infty} = 31
\end{equation}

\begin{equation}

\end{equation}

\begin{equation}
\mathbf{||x_2||_\infty} = 27
\end{equation}
