\begin{equation}
\mathbf{A} = \frac{1}{4} \begin{bmatrix}
    3 & 4 \\
    6 & 13
\end{bmatrix}
\end{equation}

**(a)** Compute the $ LU $ decomposition of it, i.e. $ \mathbf{A} = \mathbf{L}\mathbf{U} $, where $ \mathbf{L} $ is a lowertriangular matrix and $ \mathbf{U} $ is an uppertriangular matrix. Use this decomposition to solve $ \mathbf{A}\mathbf{x} = \mathbf{b} $ with $ \mathbf{b} = (1, 2)^\top $ via forward and backward substitution.

\begin{equation}
\mathbf{A} = \begin{bmatrix}
    l_{11} & 0 \\
    l_{21} & l_{22}
\end{bmatrix}
\begin{bmatrix}
    u_{11} & u_{12}\\
    0 & u_{22}
\end{bmatrix}
\end{equation}

\begin{equation}
l_{11} u_{11} + 0 \cdot 0 = 3 \\
l_{11} u_{12} + 0 \cdot u_{22} = 4 \\
l_{21} u_{11} + l_{22} \cdot 0 = 6 \\
l_{21} u_{12} + l_{22} u_{22} = 13
\end{equation}

This linear system is underdetermined. Therefore, we are going to introduce the following constraint: $ \mathbf{L} $ should be a unit triangular matrix, i.e. the diagonal elements should be 1 ($ l_{11} = 1 $ and $ l_{22} = 1 $).

Solving the linear system using the proposed constraints yields

\begin{equation}
\mathbf{L} = \begin{bmatrix}
    1 & 0 \\
    2 & 1
\end{bmatrix} \\
\mathbf{U} = \begin{bmatrix}
    3 & 4 \\
    0 & 5
\end{bmatrix}
\end{equation}

**Forward substitution**

\begin{equation}
\mathbf{A}\mathbf{x} = \mathbf{b} \\
\mathbf{L}\mathbf{U}\mathbf{x} = \mathbf{b} \\
\mathbf{L}\mathbf{y} = \mathbf{b} \\
\end{equation}

Solution:
\begin{equation}
\mathbf{y} = \begin{bmatrix}
1 \\
0
\end{bmatrix}
\end{equation}

**Backward substitution**

\begin{equation}
\mathbf{U}\mathbf{x} = \mathbf{y}
\end{equation}

Solution:
\begin{equation}
\mathbf{x} = \begin{bmatrix}
\frac{1}{3} \\
0
\end{bmatrix}
\end{equation}
