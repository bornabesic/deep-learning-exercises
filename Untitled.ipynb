## 5. Special Orthorgonal Matrices



(a)
\begin{equation}
A = \frac{1}{2}\begin{bmatrix}
7 & \sqrt{3} \\
\sqrt{3} & 5 \\
\end{bmatrix}
\end{equation}

\begin{equation}
det(A) = \frac{1}{2}(7\cdot5 - \sqrt{3}\cdot\sqrt{3}) = \frac{1}{2}(35 - 3) = 16 \\
Tr(A) = \frac{1}{2}(7 + 5) = 6
\end{equation}

Solving for eigen values
\begin{equation}
\begin{bmatrix}
7/2 & \sqrt{3}/2 \\
\sqrt{3}/2 & 5/2 \\
\end{bmatrix}
-
\begin{bmatrix}
Λ & 0\\
0 & Λ\\
\end{bmatrix}
= 0
\end{equation}

\begin{equation}
\begin{bmatrix}
7/2 - Λ & \sqrt{3}/2 \\
\sqrt{3}/2 & 5/2 - Λ \\
\end{bmatrix}
= 0
\end{equation}

$$ (7/2 - Λ)\cdot(5/2 - Λ) - \sqrt{3}/2\cdot\sqrt{3}/2 = 0 $$

$$ Λ^{2} - 6Λ + 8 = 0 $$
We get
$$ Λ = 2,4 $$
